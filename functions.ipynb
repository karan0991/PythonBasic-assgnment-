```python
#Q1  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list
def sum_even(lst):
    return sum(num for num in lst if num % 2 == 0)

print(sum_even([1, 2, 3, 4]))
```

```python
# Q3  Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
import math as m
def square_list(lst):
    return [m.pow(x,2) for x in lst]

print(square_list([1, 2, 3]))
```

```python
# Q4 Write a Python function that checks if a given number is prime or not from 1 to 200
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

primes = [num for num in range(1, 201) if is_prime(num)]

print(primes)
```

```python
# Q5  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
def fibonacci_generator(num_terms):
    a, b = 0, 1
    for _ in range(num_terms):
        yield a
        a, b = b, a + b

num_terms = 5

for number in fibonacci_generator(num_terms):
    print(number,end=" ")
```

```python
# Q6 a generator function in Python that yields the powers of 2 up to a given exponent
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

for power in powers_of_two(5):
    print(power,end=" ")
```

```python
# Q7  Implement a generator function that reads a file line by line and yields each line as a string.
def read_lines(test):
    with open(test, 'r') as f:
        for line in f:
            yield line.strip()
            
file = r"C:\Users\karan\Desktop\studyrelated things\assignments-\test.txt"
for line in read_lines(file):
    print(line)
```

```python
# Q8 Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
tuples_list = [(1, 3), (4, 1), (2, 2), (5, 0)]
sorted_list = sorted(tuples_list, key=lambda x: x[1])
print(sorted_list)
```

```python
# Q9 a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
celsius = [0, 20, 30]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print(fahrenheit)
```

```python
# Q10  Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(s):
    return ''.join(filter(lambda x: x.lower() not in 'aeiou', s))
print(remove_vowels("hello world"))
```
