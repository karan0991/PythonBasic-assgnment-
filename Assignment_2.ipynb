Q1 What is hypothesis testing in statistics
- Hypothesis testing is a statistical method used to determine whether there is enough evidence in a sample to infer that a certain condition holds for the entire population.

Q2 What is the null hypothesis, and how does it differ from the alternative hypothesis
- Null Hypothesis (H₀): Assumes no effect or no difference. Example: "The average height of students is 5.5 ft."
- Alternative Hypothesis (H₁): Assumes an effect or difference exists. Example: "The average height of students is not 5.5 ft."


Q3 What is the significance level in hypothesis testing, and why is it important
- The significance level (α) is the probability of rejecting the null hypothesis when it is true. Common levels are 0.05 or 0.01. It determines the threshold for deciding if results are statistically significant.

Q4 What does a P-value represent in hypothesis testing
-  The P-value is the probability of observing data as extreme as the sample, assuming the null hypothesis is true. Lower P-values indicate stronger evidence against H₀.

Q5 How do you interpret the P-value in hypothesis testing
- If P ≤ α Reject H₀ (evidence supports H₁).
- If P ≥ α Fail to reject H₀ (not enough evidence for H₁).

Q6 What are Type 1 and Type 2 errors in hypothesis testing
- Type 1 Error: Rejecting H₀ when it is true. Probability = α.
- Type 2 Error: Failing to reject H₀ when it is false. Probability = β.

Q7 What is the difference between a one-tailed and a two-tailed test in hypothesis testing
- One-tailed Test: Tests for an effect in one direction. Example: "Mean is greater than 5
- Two-tailed Test: Tests for an effect in both directions. Example: "Mean is not equal to 5


Q8 What is the Z-test, and when is it used in hypothesis testing
-  The Z-test compares sample means to known population means when the population variance is known and the sample size is large (n > 19).

Q9 How do you calculate the Z-score, and what does it represent in hypothesis testing
$$Z = \frac{\bar{x} - \mu}{\sigma \cdot \sqrt{n}}$$
 
- , where 
    - x̄ is the sample mean, 
    - μ is the population mean, 
    - σ is the population standard deviation, and 
    - n is the sample size. 
 - It represents how many standard deviations the sample mean is from the population mean

Q10 What is the T-distribution, and when should it be used instead of the normal distribution
-  The T-distribution is used when the sample size is small (n ≤ 30) and the population standard deviation is unknown.


Q11 What is the difference between a Z-test and a T-test
- Z-test: Requires known population variance, suitable for large samples.
- T-test: Suitable for small samples or unknown population variance.

Q12 What is the T-test, and how is it used in hypothesis testing
-  The T-test compares means between groups or a sample mean to a population mean. It is used for small samples and unknown population variance.

Q13 What is the relationship between Z-test and T-test in hypothesis testing
- Both tests evaluate means, but the T-test accounts for uncertainty in variance for small samples, while the Z-test assumes known variance.

Q14 What is a confidence interval, and how is it used to interpret statistical results
- A confidence interval provides a range of values within which the true parameter is likely to lie with a certain confidence level (e.g 98%)

Q15 What is the margin of error, and how does it affect the confidence interval
- The margin of error is the maximum expected difference between the sample statistic and the population parameter. Larger margins widen the confidence interval.

Q16 How is Bayes' Theorem used in statistics, and what is its significance
-  Bayes' Theorem calculates conditional probabilities and updates beliefs based on new evidence. It is significant in decision-making under uncertainty.

Q17 What is the Chi-square distribution, and when is it used
- The Chi-square distribution is used for categorical data to test independence or goodness of fit.


Q18 What is the Chi-square goodness of fit test, and how is it applied
-  It determines if observed frequencies match expected frequencies. Application involves calculating and comparing it to a critical value.its formula
$$\chi^2 = \sum \frac{(O - E)^2}{E}$$

Q19 What is the F-distribution, and when is it used in hypothesis testing
- The F-distribution compares variances of two samples and is used in ANOVA and F-tests.

Q20 What is an ANOVA test, and what are its assumptions
 - ANOVA (Analysis of Variance) tests if means of multiple groups are equal.
 -  its assumptions are 
    - Normality of data
    - Homogeneity of variances
    - Independence of observation


Q21 What are the different types of ANOVA tests
- One-way ANOVA: Tests one independent variable.
- Two-way ANOVA: Tests two independent variables and their interaction.

Q22 What is the F-test, and how does it relate to hypothesis testing?
- The F-test compares variances of two groups or tests multiple means in ANOVA. High F-values indicate group differences.

```python
# Q1 Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results
import numpy as np
import scipy.stats as stats

def z_test(sample, pop_mean, pop_std):
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
    return z_score, p_value

np.random.seed(42)
sample_data = np.random.normal(loc=105, scale=15, size=30)
z_stat, p_val = z_test(sample_data, pop_mean=100, pop_std=10)
print(f"Z-score: {z_stat:.5f}, P-value: {p_val:.5f}")
```

```python
# Q9 Write a Python function to calculate the confidence interval for a sample mean and explain its significance
def confidence_interval(data, confidence=0.95):

    n = len(data)  
    mean = np.mean(data) 
    std_err = stats.sem(data)

    t_value = stats.t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_value * std_err

    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound

sample_data = np.random.normal(100, 50, 30)
ci = confidence_interval(sample_data, confidence=0.95)
print(f"95% Confidence Interval: {ci}")
```

```python
# Q10 Write a Python program to calculate the margin of error for a given confidence level using sample data

def margin_of_error(data, confidence=0.95):

    n = len(data)
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    
    if n >= 30:
        critical_value = stats.norm.ppf((1 + confidence) / 2)
    else:
        critical_value = stats.t.ppf((1 + confidence) / 2, df=n-1)
    
    standard_error = sample_std / np.sqrt(n)
    
    moe = critical_value * standard_error

    return moe

np.random.seed(42)
sample_data = np.random.normal(100, 50, 30)
moe = margin_of_error(sample_data, confidence=0.95)
print(f"Margin of Error (95% confidence): {moe:.4f}")
```

```python
# Q11 Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
def bayes_theorem(prior, likelihood, false_positive):

    not_prior = 1 - prior  # P(¬H)
    marginal = (likelihood * prior) + (false_positive * not_prior)  # P(E)
    
    # Bayes' theorem: P(H|E) = (P(E|H) * P(H)) / P(E)
    posterior = (likelihood * prior) / marginal
    return posterior

prior_prob = 0.01        # P(H)
likelihood = 0.95        # P(E|H) 
false_positive = 0.05    # P(E|¬H

posterior_prob = bayes_theorem(prior_prob, likelihood, false_positive)
print(f"Posterior Probability (Having Disease | Positive Test): {posterior_prob:.4f}")

```

```python
# Q12 Perform a Chi-square test for independence between two categorical variables in Python
import pandas as pd
data = np.array([[30, 10],   # Age Group 0-10
                 [35, 25],   # Age Group 10-20
                 [20, 40]])  # Age Group 100+

categories = ["0-10", "10-20", "100+"]
brands = ["Brand A", "Brand B"]
contingency_table = pd.DataFrame(data, index=categories, columns=brands)
print("Contingency Table:\n", contingency_table)

chi2_stat, p_value, dof,expected= stats.chi2_contingency(contingency_table)

print(f"\nChi-square Statistic: {chi2_stat:.3f}")
print(f"P-value: {p_value:.3f}")
print(f"Degrees of Freedom: {dof}")


if p_value < 0.05:
    print("\nConclusion: Reject the null hypothesis. The two variables are dependent.")
else:
    print("\nConclusion: Fail to reject the null hypothesis. The two variables are independent.")
```

```python
# Q13 Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data

def calculate_expected_frequencies(observed):
 
    observed = np.array(observed)  
    row_totals = observed.sum(axis=1)  
    col_totals = observed.sum(axis=0)
    grand_total = observed.sum()  

    
    expected = np.outer(row_totals, col_totals) / grand_total

    return expected

observed_data = [
    [30, 10],   # Age Group 0-10
    [35, 25],   # Age Group 10-20
    [20, 40] # Age Group 100+
]

expected_frequencies = calculate_expected_frequencies(observed_data)

categories = ["18-25", "26-40", "41+"]
brands = ["Brand A", "Brand B"]
expected_df = pd.DataFrame(expected_frequencies, index=categories, columns=brands)

print("Expected Frequencies Table:\n", expected_df)
```

```python
# Q16 Implement an F-test using Python to compare the variances of two random samples

np.random.seed(42)

sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=50, scale=15, size=30)

var1 = np.var(sample1, ddof=1)  
var2 = np.var(sample2, ddof=1)

if var1 > var2:
    F_stat = var1 / var2
    df1, df2 = len(sample1) - 1, len(sample2) - 1
else:
    F_stat = var2 / var1
    df1, df2 = len(sample2) - 1, len(sample1) - 1

p_value = 2 * (1 - stats.f.cdf(F_stat, df1, df2))


print(f"Sample 1 Variance: {var1:.3f}")
print(f"Sample 2 Variance: {var2:.3f}")
print(f"F-statistic: {F_stat:.3f}")
print(f"P-value: {p_value:.3f}")

alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis. Variances are significantly different.")
else:
    print("Conclusion: Fail to reject the null hypothesis. Variances are equal.")

```

```python
# Q17 Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results

np.random.seed(42)

group1 = np.random.normal(loc=50, scale=10, size=30)  
group2 = np.random.normal(loc=55, scale=10, size=30)  
group3 = np.random.normal(loc=60, scale=10, size=30)

F_stat, p_value = stats.f_oneway(group1, group2, group3)

print(f"F-statistic: {F_stat:.3f}")
print(f"P-value: {p_value:.3f}")

alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis. At least one group has a significantly different mean.")
else:
    print("Conclusion: Fail to reject the null hypothesis. All group means are statistically similar.")
```

```python
# Q18 Perform a one-way ANOVA test using Python to compare the means of different groups and plot the results
import seaborn as sns

np.random.seed(42)

group1 = np.random.normal(loc=50, scale=10, size=30)  
group2 = np.random.normal(loc=55, scale=10, size=30)  
group3 = np.random.normal(loc=60, scale=10, size=30)  

F_stat, p_value = stats.f_oneway(group1, group2, group3)


print(f"F-statistic: {F_stat:.3f}")
print(f"P-value: {p_value:.3f}")


alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis. At least one group has a significantly different mean.")
else:
    print("Conclusion: Fail to reject the null hypothesis. All group means are statistically similar.")

data = {
    "Scores": np.concatenate([group1, group2, group3]),
    "Group": (["Group 1"] * len(group1)) + (["Group 2"] * len(group2)) + (["Group 3"] * len(group3))
}
df = pd.DataFrame(data)

plt.figure(figsize=(8, 5))
sns.boxplot(x="Group", y="Scores", data=df,hue="Group", palette="Set2",legend=False)

plt.title("One-Way ANOVA: Comparing Group Means", fontsize=14)
plt.xlabel("Groups", fontsize=12)
plt.ylabel("Scores", fontsize=12)
plt.grid(True)

plt.show()
```

```python
# Q19 Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

def check_anova_assumptions(groups):

    print("\n**Normality Test (Shapiro-Wilk)**")
    for i, group in enumerate(groups):
        stat, p_value = stats.shapiro(group)
        print(f"Group {i+1}: W={stat:.3f}, p={p_value:.3f}")
        if p_value < 0.05:
            print(f"  → Reject H0: Group {i+1} is NOT normally distributed.")
        else:
            print(f"  → Fail to reject H0: Group {i+1} appears normal.")

    print("\n**Equal Variance Test (Levene’s Test)**")
    stat, p_value = stats.levene(*groups)
    print(f"Levene’s Test: W={stat:.3f}, p={p_value:.3f}")
    if p_value < 0.05:
        print("  → Reject H0: Groups have significantly different variances.")
    else:
        print("  → Fail to reject H0: Variances appear equal.")

    plt.figure(figsize=(10, 4))


    plt.subplot(1, 2, 1)
    for i, group in enumerate(groups):
        sns.histplot(group, kde=True, label=f"Group {i+1}", alpha=0.5)
    plt.title("Histogram of Group Distributions")
    plt.legend()


    plt.subplot(1, 2, 2)
    sns.boxplot(data=groups)
    plt.title("Box Plot of Groups")
    plt.xticks(ticks=range(len(groups)), labels=[f"Group {i+1}" for i in range(len(groups))])

    plt.tight_layout()
    plt.show()

    print("\n**Independence Assumption**")
    print("  → Independence is assumed based on experimental design (random sampling).")

np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)
group2 = np.random.normal(loc=55, scale=10, size=30)
group3 = np.random.normal(loc=60, scale=10, size=30)

check_anova_assumptions([group1, group2, group3])
```

```python
# Q24 Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

np.random.seed(42)

sample = np.random.normal(loc=50, scale=10, size=30)  
n = len(sample)  
sample_variance = np.var(sample, ddof=1)  

sigma_0_squared = 100  

chi2_stat = (n - 1) * sample_variance / sigma_0_squared

p_value = 2 * min(stats.chi2.cdf(chi2_stat, df=n-1), 1 - stats.chi2.cdf(chi2_stat, df=n-1))


print(f"Sample Variance: {sample_variance:.3f}")
print(f"Chi-square Statistic: {chi2_stat:.3f}")
print(f"P-value: {p_value:.3f}")

alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis. The population variance is significantly different from the assumed value.")
else:
    print("Conclusion: Fail to reject the null hypothesis. No significant difference in population variance.")
```

```python
# Q25 Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

def z_test_proportions(x1, n1, x2, n2, alternative="two-tailed"):
    p1 = x1 / n1
    p2 = x2 / n2

    p_pool = (x1 + x2) / (n1 + n2)

    standard_error = np.sqrt(p_pool * (1 - p_pool) * (1/n1 + 1/n2))

    z_stat = (p1 - p2) / standard_error

    if alternative == "two-tailed":
        p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))
    elif alternative == "left-tailed":
        p_value = stats.norm.cdf(z_stat)
    elif alternative == "right-tailed":
        p_value = 1 - stats.norm.cdf(z_stat)
    else:
        raise ValueError("alternative must be 'two-tailed', 'left-tailed', or 'right-tailed'")

    return z_stat, p_value

x1, n1 = 60, 200  
x2, n2 = 50, 180

z_stat, p_value = z_test_proportions(x1, n1, x2, n2, alternative="two-tailed")
print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_value:.3f}")

alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis. The proportions are significantly different.")
else:
    print("Conclusion: Fail to reject the null hypothesis. No significant difference in proportions.")
```

```python
# Q26 Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results

np.random.seed(42)

group1 = np.random.normal(loc=50, scale=10, size=30)  
group2 = np.random.normal(loc=50, scale=15, size=30)  

var1 = np.var(group1, ddof=1)  
var2 = np.var(group2, ddof=1)

if var1 > var2:
    F_stat = var1 / var2
    df1, df2 = len(group1) - 1, len(group2) - 1
else:
    F_stat = var2 / var1
    df1, df2 = len(group2) - 1, len(group1) - 1

p_value = 2 * min(stats.f.cdf(F_stat, df1, df2), 1 - stats.f.cdf(F_stat, df1, df2))

print(f"Group 1 Variance: {var1:.3f}")
print(f"Group 2 Variance: {var2:.3f}")
print(f"F-statistic: {F_stat:.3f}")
print(f"P-value: {p_value:.3f}")

alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis. The variances are significantly different.")
else:
    print("Conclusion: Fail to reject the null hypothesis. The variances are not significantly different.")

plt.figure(figsize=(8, 5))
sns.boxplot(data=[group1, group2], palette=["blue", "red"])
plt.xticks([0, 1], ["Group 1", "Group 2"])
plt.title("Box Plot: Comparing Variances")
plt.ylabel("Values")
plt.grid()
plt.show()

plt.figure(figsize=(8, 5))
sns.histplot(group1, bins=15, color="blue", label="Group 1", alpha=0.6)
sns.histplot(group2, bins=15, color="red", label="Group 2", alpha=0.6)
plt.axvline(np.var(group1), color="blue", linestyle="dashed", label=f"Var Group 1: {var1:.1f}")
plt.axvline(np.var(group2), color="red", linestyle="dashed", label=f"Var Group 2: {var2:.1f}")
plt.title("Histogram of Variances")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid()
plt.show()
```

```python
# Q27 Perform a Chi-square test for goodness of fit with simulated data and analyze the results

categories = ["A", "B", "C", "D", "E"]
observed_counts = np.array([50, 30, 80, 40, 60])  

total_observed = sum(observed_counts)
expected_counts = np.full(len(categories), total_observed / len(categories)) 

chi2_stat, p_value = stats.chisquare(f_obs=observed_counts, f_exp=expected_counts)

print(f"Chi-square Statistic: {chi2_stat:.3f}")
print(f"P-value: {p_value:.3f}")

alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis. The observed distribution significantly differs from the expected distribution.")
else:
    print("Conclusion: Fail to reject the null hypothesis. The observed distribution matches the expected distribution.")

plt.figure(figsize=(8, 5))
x = np.arange(len(categories))
plt.bar(x - 0.2, observed_counts, width=0.4, label="Observed", color="blue", alpha=0.6)
plt.bar(x + 0.2, expected_counts, width=0.4, label="Expected", color="red", alpha=0.6)
plt.xticks(x, categories)
plt.xlabel("Categories")
plt.ylabel("Frequency")
plt.title("Chi-Square Goodness-of-Fit Test: Observed vs. Expected Frequencies")
plt.legend()
plt.grid()
plt.show()
```
